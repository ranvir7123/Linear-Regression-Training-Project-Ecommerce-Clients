```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/Ecommerce Customers')
df.head()
```

```python
df.info()
```

```python
df.drop(['Avatar','Email','Address'],axis = 1,inplace = True)
df.head()
```

```python
sns.lineplot(x='Length of Membership',
           y='Yearly Amount Spent',
           data=df)
```

```python
sns.lineplot(x='Time on Website',
           y='Yearly Amount Spent',
           data=df)
```

```python
sns.lineplot(x='Time on App',
           y='Yearly Amount Spent',
           data=df)
```

```python
from sklearn.model_selection import train_test_split
x=df.drop(['Yearly Amount Spent'],axis=1)
y = df['Yearly Amount Spent']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.32,random_state=42)


```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
predictions = lr.predict(xtest)
predictions
lr.score(x, y)
```
